```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

transformed_data = pd.read_csv("./transformed_samples_by_features.csv").drop(columns=["Unnamed: 0"])
transformed_data.head()
```

```python
level2_transform = pd.DataFrame()
column_names = transformed_data.columns

for column_index in range(len(transformed_data.columns) - 1):
    for second_column_index in range(column_index, len(transformed_data.columns)):
        level2_transform[column_names[column_index] + " Times " + column_names[second_column_index]] = transformed_data[column_names[column_index]] * transformed_data[column_names[second_column_index]]

```

```python
to_remove = []
level2_column_names = level2_transform.columns
for column_index in range(len(level2_transform.columns)):
    for second_column_index in range(len(transformed_data.columns)):
        true_false = np.where(level2_transform[level2_column_names[column_index]] == transformed_data[column_names[second_column_index]], True, False)
        if true_false[0] == True:
            to_remove.append(level2_column_names[second_column_index])
        #to_remove.append(level2_column_names[second_column_index])
level2_transform = level2_transform.drop(columns=to_remove)
level2_transform
```
