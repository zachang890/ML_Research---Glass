```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

transformed_data = pd.read_csv("./transformed_samples_by_features.csv").drop(columns=["Unnamed: 0"])
transformed_data.head()
```

```python
level2_transform = pd.DataFrame()
column_names = transformed_data.columns

for column_index in range(len(transformed_data.columns) - 1):
    for second_column_index in range(column_index, len(transformed_data.columns)):
        level2_transform[column_names[column_index] + " Times " + column_names[second_column_index]] = transformed_data[column_names[column_index]] * transformed_data[column_names[second_column_index]]

```

```python
#Algorithm behind cleaning duplicates
to_remove = []
level2_column_names = level2_transform.columns
for nontransform_column_index in range(len(transformed_data.columns)):
    for secondtransform_index in range(len(level2_transform.columns)):
        true_false = np.where(level2_transform[level2_column_names[secondtransform_index]] == transformed_data[column_names[nontransform_column_index]], True, False)
        if true_false[0] == True:
            to_remove.append(level2_column_names[secondtransform_index])

to_remove
```

```python
#Example of how np.where works
trial_df = pd.DataFrame()
trial_df["AtomicNumber Times AtomicNumber"] = level2_transform["AtomicNumber Times AtomicNumber"]
trial_df["AtomicNumber Squared"] = transformed_data["AtomicNumber"] ** 2

trial_df.at[0, "AtomicNumber Times AtomicNumber"] = 1000
true_false = np.where(trial_df["AtomicNumber Times AtomicNumber"] == trial_df["AtomicNumber Squared"], True, False)
true_false
```

```python
a = level2_transform #a is here to save the original transform just in case
level2_transform = level2_transform.drop(columns=to_remove)

# for remove in to_remove:
#     print(remove)
#     level2_transform = level2_transform.drop(columns=[remove], axis = 1)
# print(len(level2_transform.columns))
```
