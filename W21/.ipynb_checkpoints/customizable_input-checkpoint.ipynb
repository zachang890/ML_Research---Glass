```python
import pandas as pd
from mendeleev import element

class Tailored_Input:
    def __init__(self, filtered, atomic_desc, element_desc):
        #Glass samples that only use 15 of the 24 oxides
        self.filtered = pd.read_csv(filtered)
        
        #Original atomic quant descriptor file
        self.atomic_desc = pd.read_csv(atomic_desc)
        
        #Original element descriptor file
        self.element_desc = pd.read_csv(element_desc)
    def generate_matrix1(self):
        #Find which oxides to keep
        keep_oxides = [oxide for oxide in self.filtered.columns if oxide in list(self.atomic_desc["Unnamed: 0"])]
        filtered_matrix1 = pd.DataFrame()
        for row_i in range(len(self.atomic_desc)):
            if self.atomic_desc.iloc[row_i, 0] in keep_oxides:
                filtered_matrix1 = filtered_matrix1.append(pd.Series(list(self.atomic_desc.loc[row_i]), index = self.atomic_desc.columns), ignore_index=True)
        filtered_matrix1 = filtered_matrix1.loc[:, (filtered_matrix1 != 0).any(axis=0)] #filter out columns with only zero
        filtered_matrix1 = filtered_matrix1.drop("Unnamed: 0", axis=1)
        filtered_matrix1 = filtered_matrix1.reindex(sorted(list(filtered_matrix1.columns), key = lambda x: element(x).atomic_number), axis = 1)
        return filtered_matrix1
    def generate_matrix2(self, matrix1):
        #Find which elements to keep
        keep_elements = [element for element in matrix1.columns]
        filtered_matrix2 = pd.DataFrame()
        for row_i in range(len(self.element_desc)):
            if self.element_desc.iloc[row_i, 0] in keep_elements:
                filtered_matrix2 = filtered_matrix2.append(pd.Series(list(self.element_desc.loc[row_i]), index = self.element_desc.columns), ignore_index=True)
        elements = list(filtered_matrix2["Unnamed: 0"])
        print(elements)
        for row_i in range(len(filtered_matrix2)):
            filtered_matrix2 = filtered_matrix2.rename(index = {row_i: elements[row_i]})
        filtered_matrix2 = filtered_matrix2.reindex(sorted(filtered_matrix2.index, key = lambda x: element(x).atomic_number), axis = 0)
        return filtered_matrix2
```

```python
a = Tailored_Input("./pr_df_15.csv", "./atomicquant_Descriptor_table_oxide_series_4.csv", "./element_Descriptor_table_oxide_series_4.csv")
filtered_matrix1 = a.generate_matrix1()
filtered_matrix1
```
